```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is your DataFrame with the given structure

# Helper function to create a bar plot
def plot_bar(data, x_col, y_col, title, xlabel, ylabel, figsize=(12, 6), top_n=None):
    # Sort values by y_col
    data = data.sort_values(by=y_col, ascending=False)
    
    # If top_n is specified, limit the number of rows
    if top_n:
        data = data.head(top_n)
    
    # Create the bar plot
    plt.figure(figsize=figsize)
    sns.barplot(data=data, x=x_col, y=y_col, palette="viridis")
    plt.title(title, fontsize=16)
    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel(ylabel, fontsize=14)
    plt.xticks(rotation=45, ha="right")
    plt.show()

# Average premium by Gender
gender_premium_avg = df.groupby('Gender')['PremTotal'].mean().reset_index()
plot_bar(
    data=gender_premium_avg,
    x_col='Gender',
    y_col='PremTotal',
    title="Average Premium by Gender",
    xlabel="Gender",
    ylabel="Average Premium"
)

# Average premium by DrivAge
drivage_premium_avg = df.groupby('DrivAge')['PremTotal'].mean().reset_index()
plot_bar(
    data=drivage_premium_avg,
    x_col='DrivAge',
    y_col='PremTotal',
    title="Average Premium by Driver Age",
    xlabel="Driver Age",
    ylabel="Average Premium"
)

# Average premium by VehGroup
vehgroup_premium_avg = df.groupby('VehGroup')['PremTotal'].mean().reset_index()
plot_bar(
    data=vehgroup_premium_avg,
    x_col='VehGroup',
    y_col='PremTotal',
    title="Average Premium by Vehicle Group",
    xlabel="Vehicle Group",
    ylabel="Average Premium",
    top_n=10  # Show only the top 10 vehicle groups
)

# Average premium by State
state_premium_avg = df.groupby('State')['PremTotal'].mean().reset_index()
plot_bar(
    data=state_premium_avg,
    x_col='State',
    y_col='PremTotal',
    title="Average Premium by State",
    xlabel="State",
    ylabel="Average Premium",
    top_n=15  # Show only the top 15 states
)

# Average premium by Area
area_premium_avg = df.groupby('Area')['PremTotal'].mean().reset_index()
plot_bar(
    data=area_premium_avg,
    x_col='Area',
    y_col='PremTotal',
    title="Average Premium by Area",
    xlabel="Area",
    ylabel="Average Premium"
)
```

### **Step 6: Feature engineering**

**6.1 New field: Vehicle Brand**

```python
# Extract the vehicle brand name
df['VehBrand'] = df['VehGroup'].str.split(' ').str[0]
df["VehBrand"]
```

```python
# Count the occurrences of each vehicle brand
brand_counts = df["VehBrand"].value_counts()

# Create a pie chart
plt.figure(figsize=(10, 8))
plt.pie(brand_counts, labels=brand_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title("Vehicle Brand Distribution")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Calculate the average premium by vehicle brand
brand_premium_avg = df.groupby('VehBrand')['PremTotal'].mean().reset_index()

# Sort the results by the average premium in descending order
brand_premium_avg = brand_premium_avg.sort_values(by='PremTotal', ascending=False)

# Display the result
print(brand_premium_avg)

# Plot the top 10 vehicle brands by average premium
top_n = 10
plt.figure(figsize=(12, 6))
sns.barplot(
    data=brand_premium_avg.head(top_n),  # Use only the top `n` brands
    x='VehBrand',
    y='PremTotal',
    palette='viridis'
)
plt.title(f"Top {top_n} Vehicle Brands by Average Premium", fontsize=16)
plt.xlabel("Vehicle Brand", fontsize=14)
plt.ylabel("Average Premium", fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.show()
```

**6.2 Numifying DrivAge and Gender for correlation analysis**

```python
# Map drivage groups to integers
df['DrivAge_int'] = df['DrivAge'].map({'<18': 0, '18-25': 1, '26-35': 2, '36-45': 3, '46-55': 4, '>55': 5}).astype(int)

# Map gender to integers
df['Gender_int'] = df['Gender'].map({'Male': 0, 'Female': 1})
```

```python
# Select numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Plot heatmapaaa
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix (Numerical Features)")
plt.show()
```
