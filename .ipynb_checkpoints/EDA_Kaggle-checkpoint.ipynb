```python
import pandas as pd

# Load CSV file
df = pd.read_csv('C:/Users/Alan/Desktop/HKU/dataset/kaggle/aw_fb_data.csv')

# Display basic info
df.info()
```

```python
# Fix typo in column names
df.rename(columns={'hear_rate': 'heart_rate'}, inplace=True)
df.rename(columns={'entropy_setps': 'entropy_steps'}, inplace=True)
```

```python
df.shape
```

### **Step 2: Handle missing values**

```python
# Check missing values
print(df.isnull().sum())

# Drop rows if any column is empty
df = df.dropna(subset=['age','gender'])

# # Fill missing categorical values with the mode (most common value)
# categorical_cols = df.select_dtypes(include=['object']).columns
# df[categorical_cols] = df[categorical_cols].fillna(df[categorical_cols].mode().iloc[0])

# # Fill missing numerical values with median
# numerical_cols = df.select_dtypes(include=['number']).columns
# df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].median())
```

```python
# Remove unnecessary column
df.drop(columns=['Unnamed: 0'], inplace=True)
```

### **Step 3: Convert data type**

```python
df['gender'].info

# Change data type of gender from int64 to object
df['gender_type'] = df['gender'].astype('object')

# Replace the value with strings
df['gender_type'] = df['gender_type'].replace({1: 'male', 0: 'female'})
```

```python
df['gender_type']
```

### **Step 4: Summary statistics**

```python
# Summary statistics for numerical columns
print(df.describe())
```

```python
# Summary statistics for categorical columns
print(df.describe(include=['object']))
```

### **Step 5: Visualization**

**5.1 Distribution of Numerical Features**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for numerical features
df.hist(figsize=(15, 10), bins=50)
plt.show()

# Boxplot for checking outliers
plt.figure(figsize=(10, 5))
sns.boxplot(x=df["steps"])
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(x=df["distance"])
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(x=df["entropy_heart"])
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(x=df["entropy_steps"])
plt.show()
```

**5.2 Correlation Analysis**

```python
# Select numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix (Numerical Features)")
plt.show()
```

**5.3 Distribution of Categorical Features**

```python
# List of categorical columns
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
categorical_columns
```

```python
# Count the occurences of each gender
gender_counts = df["gender_type"].value_counts()

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title("Gender Distribution")
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()

# Count the occurrences of each device
device_counts = df["device"].value_counts()

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(device_counts, labels=device_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title("Device Distribution")
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()

# Count the occurrences of each activity
activity_counts = df["activity"].value_counts()

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(activity_counts, labels=activity_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title("Activity Distribution")
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

**5.4 Bivariate analyses**

```python
# Calculate average heart rate and calories by age group
average_metrics = df.groupby(['age_group','activity']).agg({
    'heart_rate': 'mean',
    'resting_heart': 'mean',
    'calories': 'mean'
}).reset_index()

average_metrics
```

```python
# Count device types by age group
device_counts = df.groupby(['age_group', 'device']).size().unstack(fill_value=0)
device_counts
```

### **Step 6: Feature engineering**

```python
# Bucket the age into groups
bins = [-1, 17, 25, 35, 45, 55, float('inf')]  # Define age ranges
labels = ['<18', '18-25', '26-35', '36-45', '46-55', '>55']  # Corresponding labels

# Convert age to categorical groups
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, ordered=True)
```

```python
df[['age_group','age']]
```

```python
# Count the occurrences of each Age Group
age_group_counts = df["age_group"].value_counts()

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(age_group_counts, labels=age_group_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title("Age Group Distribution")
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()

# Calculate percentages
total_count = age_group_counts.sum()
age_group_percentages = (age_group_counts / total_count) * 100

# Print age groups and their percentages
for age_group, percentage in zip(age_group_counts.index, age_group_percentages):
    print(f"{age_group}: {percentage:.2f}%")
```
